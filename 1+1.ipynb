```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
import random as r
```

```python

rozrzut=10
wsp_przyrostu=1.1
l_iteracji=50

zakres_zmiennosci = np.arange(0, 101)
x = zakres_zmiennosci[r.randint(0, 101)]
y = np.sin(x / 10) * np.sin(x / 200)

for i in range(l_iteracji):
    plt.figure(figsize=((12,8)))
    while True:
        x_pot = x + r.uniform(-rozrzut, rozrzut)
        if 0 <= x_pot <= 100:
            break
    y_pot = np.sin(x_pot / 10) * np.sin(x_pot / 200)
    if y_pot >= y:
        x = x_pot
        y = y_pot
        rozrzut *= wsp_przyrostu
    else:
        rozrzut /= wsp_przyrostu
    print(f'iteracja-{i}, x={round(x,3)}, y={round(y,3)}, rozrzut={rozrzut}')
    print()

    wynik = np.sin(zakres_zmiennosci / 10) * np.sin(zakres_zmiennosci / 200)

    plt.plot(zakres_zmiennosci, wynik, label='zakres zmiennosci')
    plt.plot(x, y, 'ro', label='f(x)')
    plt.legend()
    plt.show()
```
